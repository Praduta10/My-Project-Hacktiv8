# IMPORT PUSTAKA YANG DIBUTUHKAN

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

# DATA LOADING

Pada bagian ini data loading dimasukkan beberapa rumus seperti gradien magnitude dan gradien turunan parsial pada gamabar

```python
img = Image.open("201030094143-stock-rhodesian-ridgeback-exlarge-169.jpg")
img
```

```python
arrayimg = np.array(img)
arrayimg
```

```python
#Penggunaan gradien magnitude pada gambar anjing
arrayimg_x = np.gradient(arrayimg, axis = 0)
arrayimg_y = np.gradient(arrayimg, axis = 1)
```

```python
mag = np.sqrt(((arrayimg_x)**2 + (arrayimg_y)**2))
mag
```

```python
mag.shape
```

```python
#Penggunaan transform Linear pada gambar anjing
image_flip = img.transpose(Image.FLIP_LEFT_RIGHT)
image_mirror = np.fliplr(img)
```

```python
fig = plt.figure(figsize = (10,20))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

ax1.title.set_text("Gambar Cermin Dari Anjing")
ax2.title.set_text("Gambar Flip Dari Anjing")
ax1.imshow(image_mirror)
ax2.imshow(image_flip)
```

# IMAGE PROCESSING

Saya akan membdaningkan nilai magnitude dengan nilai threshold , dan mengukur threshold yang cocok untuk gambar anjing tersebut:

```python
thresh1 = 0
thresh2 = 15
thresh3 = 30
thresh4 = 45
thresh5 = 60


thresh1array = np.where(mag > thresh1, 255, 0)
thresh2array = np.where(mag > thresh2, 255, 0)
thresh3array = np.where(mag > thresh3, 255, 0)
thresh4array = np.where(mag > thresh4, 255, 0)
thresh5array = np.where(mag > thresh5, 255, 0)
```

```python
#Jika Threshold = 0
img_green0 = np.zeros(thresh1array.shape, dtype="uint8")
img_green0[:,:,1] = thresh1array[:,:,1]
image_img_green = Image.fromarray(img_green0)
image_img_green
```

```python
#Jika threshold = 30
img_green30 = np.zeros(thresh3array.shape, dtype="uint8")
img_green30[:,:,1] = thresh3array[:,:,1]
image_img_green = Image.fromarray(img_green30)
image_img_green
```

```python
#Jika threshold = 45
img_green45 = np.zeros(thresh4array.shape, dtype="uint8")
img_green45[:,:,1] = thresh4array[:,:,1]
image_img_green = Image.fromarray(img_green45)
image_img_green
```

```python
#Jika threshold = 60
img_green60 = np.zeros(thresh5array.shape, dtype="uint8")
img_green60[:,:,1] = thresh5array[:,:,1]
image_img_green = Image.fromarray(img_green60)
image_img_green
```

```python
#Jika threshold = 15
img_green15 = np.zeros(thresh2array.shape, dtype="uint8")
img_green15[:,:,1] = thresh2array[:,:,1]
image_img_green = Image.fromarray(img_green15)
image_img_green
```

# HASIL DAN KESIMPULAN

Dalam Setup Threshold dapat diambil kesimpulan bahwa :
    
- Array pada pixel gambar anjing berlaku threshold semakin tinggi maka edge semakin tidak terdeteksi
- Semakin rendah Threshold maka akan semakin terlihat edge yang akan dicari, namun noise akan semakin besar
- Jika threshold menemukan angka yang pas diantara threshold (10-15) maka pendeteksian gambar pada tepi bisa terlihat pada 
  gambar anjing di atas ini.
- Jika threshold diantara angka (30-45) maka garis-garis pada gambar akan menjadi warna hitam tetapi masih terlihat gambar         anjing dan sudah tidak menemukan pixel gambar pada tepi lagi.


Jadi threshold yang pas pada gambar anjing disekitaran threshold pada angka 10-15.
